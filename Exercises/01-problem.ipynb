#### After so many attempts (using filtering boolean), I had to replace the elements that had '-' with the None type and thus calculate some statistics with [replace()](https://www.geeksforgeeks.org/python-pandas-dataframe-replace/) function.

````python
covid_india_df.ConfirmedIndianNational = covid_india_df.ConfirmedIndianNational.replace('-', None) # Replacing data that contains '-'
covid_india_df.iloc[:,4:5].describe() # Making describe just a column
````

````python
print("The min of ['ConfirmedIndianNational'] column is: ",covid_india_df['ConfirmedIndianNational'].min())
print("The max of ['ConfirmedIndianNational'] column is: ",covid_india_df['ConfirmedIndianNational'].max())
````

#### If you try to get std () of that column, an error will happen (type error, because all numbers are strings), You need to convert to a numeric type and so You can calculate the standard deviation.
- [Solution: ](https://datatofish.com/string-to-integer-dataframe/)

````python
covid_india_df['ConfirmedIndianNational'] =  pd.to_numeric(covid_india_df['ConfirmedIndianNational']) # Converting
print("The standart deviation of ['ConfirmedIndianNational'] column is: ",covid_india_df['ConfirmedIndianNational'].std())
````
